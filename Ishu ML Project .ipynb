```python
## Preparing the Dataset
```

```python
dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
dataset['Price']=boston.target
```

```python
## dividing independent and dependent features
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
## Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
## Standardize the dataset (because for gradient descent to reach the global minima it is important that all of our independent features should be in same scale for good convergence)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
## Model Training
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which parameter the model has been trained
regression.get_params()
```

```python
## Prediction with test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
## Assumptions
```

```python
## Plot a scater plot for prediction
plt.scatter(y_test,reg_pred)
```

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
## plot the residuals
sns.displot(residuals,kind="kde")
```

```python
# Scatter plot with respect to prediction and residuals
# uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
# Adjusted R square and R square
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
#display adjusted r square
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
